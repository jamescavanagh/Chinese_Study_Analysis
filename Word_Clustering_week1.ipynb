```python
vocabTable = pd.read_csv('vocabList.csv',index_col=0)
```

```python
#tokenizer = AutoTokenizer.from_pretrained("ckiplab/albert-base-chinese")

model = SentenceTransformer("ckiplab/albert-base-chinese",device='cpu')
```

```python

#inputs    = tokenizer(sentence, return_tensors="pt").to(device)
#model     = model.to(device)
#outputs   = model(**inputs)
```

```python
corpus = list(vocabTable['Characters'])

corpus_embeddings = model.encode(corpus)
```

```python

# Perform kmean clustering
num_clusters = 8
clustering_model = KMeans(n_clusters=num_clusters)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_

clustered_sentences = [[] for i in range(num_clusters)]
for sentence_id, cluster_id in enumerate(cluster_assignment):
    clustered_sentences[cluster_id].append(corpus[sentence_id])

```

```python
clusterKeys = {}

for i, cluster in enumerate(clustered_sentences):
    #print("Cluster ", i+1)
    keys = dict(zip(cluster, [i] * len(cluster)))
    clusterKeys.update(keys)
```

```python
vocabTable['cluster'] = vocabTable['Characters'].map(clusterKeys)
```

```python
vocabTable.sort_values('cluster',inplace=True)
```
